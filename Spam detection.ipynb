```python
import numpy as np #for work with numeric data in dataset
import pandas as pd #for manupulate and analysis the dataset
```

```python
df=pd.read_csv('spam.csv', encoding='latin1') # TO read the csv file
df.head()  #to show the 1st 5 row
```

```python
df.isnull().sum()     # to check how many NaN exist in a spacific row
```

```python
# drop the NaN column
df.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1,inplace=True)     #drop the NaN column
df.head()
```

```python
#Convert text to features using CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
x=cv.fit_transform(df['v2'])
```

```python
print(cv.vocabulary_)   # Frequency of the existing word (word repeating time )
```

```python
from sklearn.preprocessing import LabelEncoder     #using label encoder to covert catagorical data into numerical
le=LabelEncoder()
df['v1']=le.fit_transform(df['v1'])     #covert v1 column into numerical
```

```python
le.classes_ # how many classes are exist in v1
```

```python
#take v1 as a output variable
y=df['v1'].values
y
```

```python
from sklearn.model_selection import train_test_split     #split the dataset into 80 ,20 ratio for training and testing
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#import naive base model
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(xtrain,ytrain)
```

```python
ypred=model.predict(xtest)   #predict the y using testing data
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print(confusion_matrix(ytest,ypred))
print(classification_report(ytest,ypred))
accuracy_score(ytest,ypred)
```

For 80% ,20% training testing data prdiction has a nominal error ,and the accuracy score is 98%
